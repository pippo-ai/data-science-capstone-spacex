Next, let's create markers for all launch records.

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# Closest coastline point (use MousePosition to pick yours)
# Find coordinate of the closest coastline (picked with MousePosition)
coastline_lat = 28.56367
coastline_lon = -80.57163
# Selected launch site: CCAFS LC-40
launch_site_lat = 28.562302
launch_site_lon = -80.577356


distance_coastline = calculate_distance(
    launch_site_lat, launch_site_lon,
    coastline_lat, coastline_lon
)

distance_coastline
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
coastline_coordinate = [coastline_lat, coastline_lon]

distance_marker = folium.Marker(
    coastline_coordinate,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' %
             "{:10.2f} KM".format(distance_coastline),
    )
)

site_map.add_child(distance_marker)
site_map
```

### Proximity Analysis Results

**Are launch sites in close proximity to railways?**

**Are launch sites in close proximity to highways?**  
Yes. Highways are visible near the launch sites, which facilitates ground transportation and operational logistics.

**Are launch sites in close proximity to coastline?**  
Yes. The launch site is very close to the coastline (approximately **0.58 km**), which is ideal for safety reasons and allows launch trajectories over the ocean.

**Do launch sites keep certain distance away from cities?**  
Yes. The launch site is located at a considerable distance from the closest city (approximately **14.03 km**), helping to reduce risks related to noise, safety, and potential launch failures.

**Explanation:**  
Launch sites are strategically positioned close to transportation infrastructure such as railways and highways to support efficient logistics. They are also located very close to the coastline to enable safe launches over the ocean. At the same time, a significant distance from cities is maintained to minimize risks to populated areas and ensure public safety.
